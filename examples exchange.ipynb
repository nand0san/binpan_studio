# Top gainers of the last 24 hours can be requested from exchange handler

```python
top = handlers.exchange.get_top_gainers(decimal_mode=False, api_key=api_key, api_secret=api_secret)
# this can take a while
top
```

# Most traded in some quote

```python
data = handlers.exchange.get_top_gainers(decimal_mode=True, api_key=api_key, api_secret=api_secret, margin_required=True, drop_legal=True, my_quote='USDT', sort_by_column="USDT_volume")
```

```python
data
```
