# Export and Import CSV Data

BinPan allows you to save kline (candlestick) data, atomic trades, and aggregated trades to CSV files.
These files can later be re-imported to create a `Symbol` instance without making new API requests.

```python
sym = binpan.Symbol(symbol="ltcbtc", tick_interval="1m")
sym.df
```

```python
sym.save_csv()
```

## Export Trades to CSV

Both atomic trades and aggregated trades can be exported with `save_atomic_trades_csv()` and `save_agg_trades_csv()` respectively.

```python
sym.get_atomic_trades()
```

```python
sym.save_atomic_trades_csv()
```

```python
sym.get_agg_trades()
```

```python
sym.save_agg_trades_csv()
```

## Import Klines from CSV

Use `from_csv` with the filename to create a `Symbol` instance from a previously saved CSV file.
Passing `from_csv=True` (boolean) shows an interactive file selection menu — useful in console but
may not work in all Jupyter environments.

```python
# Build the filename from the Symbol metadata (same pattern as save_csv)
start, end = sym.get_timestamps()
klines_csv = f"{sym.df.index.name.replace('/', '-')} klines {start} {end}.csv"

tt = binpan.Symbol(from_csv=klines_csv)
tt.df
```

## Import Trades from CSV

Similarly, you can load aggregated or atomic trades from CSV using the `from_csv=True` parameter in `get_agg_trades()` or `get_atomic_trades()`.

```python
# Build the agg trades filename
agg_csv = f"{sym.agg_trades.index.name.replace('/', '-')} aggTrades {start} {end}.csv"
tt.get_agg_trades(from_csv=agg_csv)
tt.agg_trades
```

```python
# Build the atomic trades filename
atomic_csv = f"{sym.atomic_trades.index.name.replace('/', '-')} atomicTrades {start} {end}.csv"
tt.get_atomic_trades(from_csv=atomic_csv)
tt.atomic_trades
```

## Verify Imported Data

After importing from CSV, the `Symbol` instance contains the same DataFrames as if fetched from the API. You can plot, add indicators, or run analysis as usual.

```python
tt.plot_agg_trades_size()
```

## Cleanup

Remove the CSV files generated during this tutorial.

```python
import os

for csv_file in [klines_csv, agg_csv, atomic_csv]:
    if os.path.exists(csv_file):
        os.remove(csv_file)
        print(f"Deleted: {csv_file}")
```
