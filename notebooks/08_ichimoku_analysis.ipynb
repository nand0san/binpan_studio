# Ichimoku Interpretations

## 1. Price Passing Through and Then Over the Cloud

**Bullish signal.** Price crosses through the cloud and breaks out above it. This indicates a shift from uncertainty to bullish momentum.

```python
sym.plot(zoom_start_idx=380, zoom_end_idx=420, volume=False)
```

## 2. Price Is Over the Cloud

**Support.** The upper edge of the cloud acts as a support level. As long as price remains above the cloud, the trend is considered bullish.

```python
sym.plot(zoom_start_idx=400, zoom_end_idx=450, volume=False)
```

## 3. Price Passing Through and Then Under the Cloud

**Bearish signal.** Price crosses through the cloud and breaks below it, indicating bearish momentum.

```python
sym.plot(zoom_start_idx=100, zoom_end_idx=200, volume=False)
```

## 4. Price Is Underneath the Cloud

**Resistance.** The lower edge of the cloud acts as resistance. While price remains below the cloud, the trend is considered bearish.

```python
sym.plot(zoom_start_idx=330, zoom_end_idx=350, volume=False)
```

## 5. Price Located Inside the Cloud

**Uncertainty / Indecision zone.** When price is inside the cloud, there is no clear trend. It is best to wait for a breakout before making trading decisions.

```python
sym.plot(zoom_start_idx=250, zoom_end_idx=300, volume=False)
```

## 6. Kumo Twist (Bearish/Bullish Cloud Switch)

**Trend reversal signal.** When the leading cloud (Senkou Span A and B, projected 26 periods ahead) crosses from green to red, it indicates a bearish Kumo twist. The reverse (red to green) indicates a bullish Kumo twist.

```python
sym.plot(zoom_start_idx=280, zoom_end_idx=350, volume=False)
```

## 7. Kumo Depth Analysis

**Volatility and trend strength filter.** A thicker cloud (greater depth between Senkou Span A and B) suggests stronger support/resistance and indicates that there was significant volatility 26 periods before.

A thin cloud suggests weaker support/resistance and a potential area where price can break through more easily.

```python
sym.plot(zoom_start_idx=150, zoom_end_idx=250, volume=False)
```
