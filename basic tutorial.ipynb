but it is recommended not to enable trading capability in the Binance API key configuration, just for your own peace of mind.

Be careful out there!

# Get data

Select a symbol, a tick interval, and a time interval -- or just a limit of candles.

Time interval inputs can be a string like `2022-02-14 12:00:00`.

```python
# Fetch the last 200 candles for LTCBTC, 1-minute interval
lun = binpan.Symbol(symbol='LTCUSDT',
                    tick_interval='1m',
                    limit=200,
                    time_zone='Europe/Madrid',
                    closed=True)
lun.df
```

```python
list(lun.df.columns)
```

# Trades

Trades can be downloaded for the time period of the instance.

**Beware of requesting large time periods.**

BinPan monitors the API weight and waits if it is approaching the limit.

```python
lun.get_atomic_trades()
```

```python
lun.get_agg_trades(minutes=60)
```

# Plots with trades

Using `plot_agg_trades_size()`, you can visualize the size of aggregated trades. Useful for finding whale operations or support/resistance levels.

```python
lun.plot_agg_trades_size()
```

# Support and resistance levels with trades
