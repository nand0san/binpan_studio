# Database Info

The `Database` class connects to a PostgreSQL/TimescaleDB instance to query stored market data. It provides methods to list tables, check row counts, inspect columns, and retrieve database statistics.

> **Note:** This notebook requires a running PostgreSQL/TimescaleDB server. Database credentials (`postgresql_user`, `postgresql_password`, `postgresql_database`, `postgresql_port`, `postgresql_host`) are loaded from `secret.py` by default, but can also be passed as arguments to the `Database` constructor.

## Connecting to the Database

```python
db = binpan.Database(host="192.168.89.211")
```

## Exploring Tables

```python
tables = db.get_tables(table_type="kline_1m")
tables
```

```python
table_counts = db.get_table_counts()
table_counts
```

```python
ser = table_counts.sort_values(ascending=False)
ser
```

```python
table = ser.index[0]
table
```

```python
db.get_columns(table_name=table)
```

```python
symbol = table.split("_")[0]
tick_interval = table.split("_")[-1]
```

```python
sym = binpan.Symbol(symbol=symbol,
                    tick_interval=tick_interval,
                    time_zone='Europe/Madrid',
                    hours=8766,
                    postgres_klines="192.168.89.211")
```

```python
type(sym.df.index)
```

```python
sym.discontinuities
```

```python
sym.repair_continuity()
```
