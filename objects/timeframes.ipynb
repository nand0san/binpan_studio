```python
from timeframes import Timestamp, parse_timestamp, Timeframe
from datetime import datetime, timezone, timedelta
import pytz
```

### Pruebas con Parseos

```python
dt = parse_timestamp("2024-01-27 03:14:00", None)
print(dt, dt.tzinfo)
dt = parse_timestamp("2024-01-27 03:14:00+01:00", None)
print(dt, dt.tzinfo)
dt = parse_timestamp("01-01-2024 3:14:00", "Europe/Madrid")
print(dt, dt.tzinfo)
```

```python
dt = parse_timestamp("2024-01-27 03:14:00+01:00", "Europe/Madrid")
print(dt, dt.tzinfo)
dt = parse_timestamp("2024-01-27 03:14:00+01:00", pytz.timezone("Europe/Madrid"))
print(dt, dt.tzinfo)
```

```python
dt = parse_timestamp("01-01-2024", None)
print(dt, dt.tzinfo)
```

### Instanciando desde datetime

```python
offset = timedelta(hours=1)
simple_timezone = timezone(offset)

# Crea un objeto datetime "aware" utilizando el desplazamiento de zona horaria definido anteriormente.
aware_datetime = datetime(2024, 1, 27, 3, 14, 15, tzinfo=simple_timezone)
aware_datetime
```

```python
aware_datetime.tzinfo
```

```python
# aware_datetime.tzinfo.zone
```

```python
# aware_datetime.tzinfo.DstTzInfo
```

```python
# instanciamos con datetime aware
dt_a = Timestamp(aware_datetime, None)
print(dt_a.timezone, type(dt_a.timezone))
dt_a
```

```python
Timestamp(aware_datetime, dt_a.timezone)
```

```python
dt_b = Timestamp(aware_datetime)
dt_b
```

```python
dt_a == dt_b
```

```python
dt_a - dt_b.ms
```
