# Esquema de objetos

Esquema General del Sistema
 
- ApiClient: Clase base para la comunicación HTTP.
- ApiProvider (Interfaz): Define operaciones comunes para proveedores de API.
- BinanceApiProvider: Implementa ApiProvider para Binance.
- IDataFetcher, IDataProcessor, IDataStorage (Interfaces): Definen operaciones específicas de obtención, procesamiento y almacenamiento de datos.
- DataHandler: Coordina el flujo de datos usando las interfaces definidas.


Esquema de Diseño Propuesto

- ApiClient: Base para la comunicación HTTP, utilizando ApiSecurityManager para asegurar las solicitudes y ApiResponseHandler para procesar las respuestas.
- ApiSecurityManager: Gestiona la seguridad de las solicitudes/respuetas.
- ApiResponseHandler: Analiza y maneja las respuestas de la API.
- ApiConfigManager: Almacena y gestiona la configuración necesaria para las diferentes APIs.
- ApiLogger: Registra las operaciones de la API para monitoreo y depuración.
- ApiCacheManager: Opcionalmente cachea respuestas para mejorar el rendimiento.
