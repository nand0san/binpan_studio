# Exercise 2.1.c
Serial correlation of returns.

```python
def autocorrelation_coefficient_(data: pd.DataFrame):
    """
    Check serial correlation of return.
    """ 
    df = data.copy(deep=True)
    
    # calculate the returns
    df['Return'] = df['Close'].pct_change()

    corr = df['Return'].autocorr()
    
    print(f"Correlation: {corr}")
```

```python
autocorrelation_coefficient(tb)
```

```python
autocorrelation_coefficient(vb)  # best
```

```python
autocorrelation_coefficient(db)
```

# Exercise 2.1.d
Compute variance of returns for every bar type. Compute the variance of those variances. In hourly basis.

```python
def compute_variances_(data: pd.DataFrame):
    """
    Hourly resampled bars variance and variance of variances.
    """
    
    df = data.copy(deep=True)

    # calculate the returns
    df['Return'] = df['Close'].pct_change()

    # group the data by hour and calculate the variance of returns for each hour
    hourly_returns = df['Return'].groupby(pd.Grouper(freq='H')).var()

    # calculate the variance of the hourly variances
    hourly_variances = hourly_returns.var()

    return hourly_returns, hourly_variances
```

```python
compute_variances(tb)
```

```python
compute_variances(vb)  # best
```

```python
compute_variances(db)
```

# Exercise 2.1.e
Compute variance of returns for every bar type. Compute the variance of those variances. In hourly basis.

```python
from scipy.stats import jarque_bera

def jarque_bera_test_(data):
    
    df = data.copy(deep=True)

    # calculate the returns
    df['Return'] = df['Close'].pct_change()

    # drop the first row, which will have a NaN return
    df = df.dropna()

    # perform the Jarque-Bera test on the returns
    test = jarque_bera(df['Return'])
    return test
```

```python
jarque_bera_test(tb)
```

```python
jarque_bera_test(vb)  # best
```

```python
jarque_bera_test(db)
```
